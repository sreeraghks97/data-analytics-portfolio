Load the e-commerce transaction dataset into a Pandas DataFrame and perform an initial assessment to prepare the data for exploratory analysis.

```python
len(ecom[
    (ecom["CC Provider"] == "American Express") &
    (ecom["Purchase Price"] > 95)
])
```

**Identify customers with credit cards expiring in 2025 to support card renewal and payment information update campaigns.**

```python
len(ecom[
    (ecom["CC Exp Date"] > "12/24") &
    (ecom["CC Exp Date"] < "01/26")
])
```

```python
len(ecom[ecom["CC Exp Date"].str.endswith("25")])
```
