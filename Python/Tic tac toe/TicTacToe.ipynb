```python
import random

# FUNCTIONS

def display_board(board):
    print(board[7] + '|' + board[8] + '|' + board[9])
    print('-+-+-')
    print(board[4] + '|' + board[5] + '|' + board[6])
    print('-+-+-')
    print(board[1] + '|' + board[2] + '|' + board[3])


def place_marker(board, marker, position):
    board[position] = marker


def win_check(board, mark):
    if mark == board[1] and mark == board[2] and mark == board[3]:
        return True
    if mark == board[4] and mark == board[5] and mark == board[6]:
        return True
    if mark == board[7] and mark == board[8] and mark == board[9]:
        return True
    if mark == board[1] and mark == board[4] and mark == board[7]:
        return True
    if mark == board[2] and mark == board[5] and mark == board[8]:
        return True
    if mark == board[3] and mark == board[6] and mark == board[9]:
        return True
    if mark == board[1] and mark == board[5] and mark == board[9]:
        return True
    if mark == board[3] and mark == board[5] and mark == board[7]:
        return True

    return False


def space_check(board, position):
    return board[position] == ' '


def full_board_check(board):
    for num in range(1, 10):
        if space_check(board, num):
            return False
    return True


def choose_first():
    if random.randint(0, 1) == 0:
        return 'Player 1'
    else:
        return 'Player 2'


def player_input():

    player1 = ''

    while player1 not in ['X', 'O']:

        try:
            player1 = input("Please choose X or O: ").upper()
        except:
             print("Please enter numbers only.")

        if player1 not in ['X', 'O']:
            print("Invalid entry, try again.")

    if player1 == 'X':
        player2 = 'O'
    else:
        player2 = 'X'

    print(f"Player 1 is {player1}, Player 2 is {player2}")

    return player1, player2


def player_choice(board):
    position = 0

    while position not in range(1, 10) or not space_check(board, position):

        try:
            position = int(input("Select the position (1-9): "))
        except:
            print("Invalid Entry")

        if position not in range(1, 10):
            print("Please enter a number between 1 and 9.")

        elif not space_check(board, position):
            print("That position is already occupied.")

    return position


def replay():
    choice = input("Do you want to play again? Enter Yes or No: ").upper()

    if choice == "YES":
        return True
    else:
        return False


# MAIN PROGRAM 

print("Welcome to Tic Tac Toe!")

while True:

    board = [' '] * 10

    player1_marker, player2_marker = player_input()

    turn = choose_first()

    print(turn + " will go first.")

    play_game = input("Are you ready to play? Enter Yes or No: ")

    if play_game.upper() == 'YES':
        game_on = True
    else:
        game_on = False
        break

    while game_on:

        if not full_board_check(board):

            if turn == 'Player 1':

                display_board(board)

                position = player_choice(board)

                place_marker(board, player1_marker, position)

                game_won = win_check(board, player1_marker)

                if game_won:
                    print("Player 1 won")
                    display_board(board)
                    game_on = False
                else:
                    turn = 'Player 2'

            elif turn == 'Player 2':

                display_board(board)

                position = player_choice(board)

                place_marker(board, player2_marker, position)

                game_won = win_check(board, player2_marker)

                if game_won:
                    print("Player 2 won")
                    display_board(board)
                    game_on = False
                else:
                    turn = 'Player 1'

        else:
            print("Game Draw")
            game_on = False

    if not replay():
        break
```
